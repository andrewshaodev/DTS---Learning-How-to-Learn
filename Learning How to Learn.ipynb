Two Fundamental Ways of Thinking:

* Focused:
* Diffused:

**Procrastination:**

When you look at something that you really rather not do, it seems that you activate the areas of your brain associated with pain

Your brain, naturally enough, looks for a way to stop that negative stimulation by switching your attention to something else.

Researchers discovered that not long after people might start actually working out what they didn't like, that neurodiscomfort disappeared

**Why Can Math and Science Be Difficult to Understand?**

Unlike a cow in a field to represent the concept of a cow, there is no analoguous concept of a cow

**Ways to Solve this Issue: Practice Makes Permanent**

Practicing sometimes many times gradually will stengthen the neural connections

Taking a break after a certain time (pomodoro) allows diffuse learning to reinforce the concepts, even though it seems like simple relaxation

Pomdoro is useful against procrastination

**Memory:**

Long Term Memory:
* A Storage Warehouse (or persistent disk)
* Distributed across different regions of the brain
* To store and item of information in long term memory you need to revisit it at least several times
* Stores fundamental techniques and concepts of what you are learning about

Working Memory:
 * What you are immediately and consciously processing in your mind
 * Widely believed that we can hold four "chunks" of working memory
 * Humans tend to group items into chunks so it seems our working memory is bigger than it actually is
 * Need to constantly repeat items to keep in working memory
 * Shutting your eyes reduces other items that allows you to keep Working Memory items in place


 Spaced Repetition Techinque:

## Week 2: Chunking and Seeing the Bigger Picture

### Chunking

Chunking is the mental leap that helps you unite bits of information together through meaning

The new logical whole makes the chunk easier to remember, and also makes it easier to fit the chunk into the larger picture of what you're learning

Just memorizing a fact without understanding or context doesn't help you understand what's really going on or how the concept fits together with other concepts you're learning

Example: Chunks are combined individual letters with each other to create words: C, O, W versus the word cow

Chunking helps your brain run more efficiently (actively encouraging your brain to chunk can be useful)

The octopus combines the four pieces of accessible working memory with long term memory for processing (I/O?):

Screenshot 2023-07-14 111152.png

When you are stressed, your attentional octopus begins to lose the abillity to make some of the connections

**How to Form a Chunk:**

The ultimate goal is to be able to find connections between the chunks your brain is chunking together into a bigger whole, to understand a subject more holistically

You may need to begin with several mini chunks, for Example, catching on to the patterns of parts of a song, instead of the whole song when trying to play it on an instrument

**Steps to Chunk Formation**

1. Focus your undivided attention on the information you want to Chunk
2. Understand the basic idea of what you are trying to chunk
3. The third step to chunking is gaining context, so you can see not just how but also when to use this chunk.
  * This allows you to be able to understand when to use the new information you gained, and also when *not* to use it

**Illusions of Competence**

Recalling: After first absorbing the material, look away and try to recall what you just absorbed.

Just absorbing the material passively, or seeing the solution to a problem without trying to do it for yourself gives an illusion of learning, without being able to master it

Testing yourself is a form of recall, making sure you understand the concepts you are trying to learn.

### Bigger Picture

**Why is it easier to learn when we are interested/motivated in the subject matter?**

Acetylcholine neurons form neuromodulatory connections to the cortex that are particularly important for focused learning when you are paying close attention.

Dopamine: Dopamine is released when we receive an unexpected reward. Dopamine is unreleased not just for immediate rewards, but for expected future rewards.

Serotonin: Has strong effects on your social life. Low Serotonin increases risk taking activity (Criminals in prison have very low levels of serotonin)

**Building a Library of Chunks:**

People enhance their knowledge and expertise by gradually building chunks in their mind

The more Chunks in your mental library, whatever subject your learning, the more easily you'll be able to solve problems in other fields

Transfer: The ability to relate one Chunk to another, not necessarily in the same field, but also in other fields. This allows a shortcut to understanding new ideas.

Example: Recognizing that key value pairs are everywhere in software: JSON, Configuration, IAM, Key Value Databases. This allows you to take your understand of key value to different pieces of software that follow the same pattern.



**Forms of Problem Solving:**
1. Sequential: where each small step leads deliberately towards a solution, involves the focused mode.
2. Holistic (global): creative diffuse mode linking of several seemingly different focused mode thoughts.

Most difficult problems and concepts are grasped through intuition, because these new ideas make a leap away from what you're familiar with.

The diffuse, semi-random way of making connections means that the solutions it provides should be very carefully verified using the focused mode. Intuitive insights aren't always correct.

**Overlearning, Choking, Einstellung, Chunking, and Interleaving**

**Overlearning:**

Repetition has its place, but overlearning can be a waste of time, especially for something that makes you anxious/nervous.

However, once you understand the basic idea, it is advantageous from a learning standpoint to move on.

Constantly focusing on what you have already learned may force you to only see one solution to an issue.

You may also cause the Illusion of Competence

**Einstellung:**

A strongly developed neurual pattern that is already strongly developed may overshadow new ideas that may lead to better outcomes.

Focused modes may prevent you from reaching the more effective or only real solution.

You sometimes need to unlearn a misconception that is already strongly formed in your mind

**Interleaving:**

Mastering a new subject means learning not only the basic chunks, but also learning how to select and use different chunks.

The best way to learn that is by practicing jumping back and forth between problems or situations that require different techniques or strategies.

Example:

In science and math in particular it can help to look ahead at the more varied problem sets that are sometimes found at the end of chapters.

Or you can deliberately try to make yourself occasionally pick out why some problems call for one technique as opposed to another.

Reinforce your brain that just knowing how to use a techinque is not enough, you also need to use it for the right context.



## Week 3: Procrastination and Memory

### Procrastiation

Handling procrastination is important for learning because it is a bit by bit activity. If you procrastinate until the last minute, you will do yourself a disservice to your learning if you try to cram

While procrastination is easy to invoke, willpower is difficult. There are ways to avoid procrastination without needing to spend your mental resources channeling your willpower

Your brain will create poor rationalizations to help you procrastinate

**Making the Conscious into Unconscious:**

Habits can be seen as energy savers for us, they free our mind for other more challenging activities

Habits is related to chunking

Habits have forr parts:

1. The Cue: The trigger that launches you into "zombie mode"
2. The Routine: The zombie mode: the routine habitual response your brain is used to falling into when it receives the cue. Can be helpful, harmless, or harmful
3. The Reward: What we get at the end of the routine. Procrastination as the routine is easy because there is instant gratification from doing something more pleasant to you.
 * Finding ways to reward yourself for good habits is a good way to escape procrastination
4. The Belief: To change a habit you want to change, you need to change your underlying belief

**Process vs Product:**

Focusing on Process vs Product may be helpful as a way to reframe your mind against procrastination

Process: the flow of time and the habits associated with that flow of time. Example: I will spend 25 minutes working on a problem instead of thinking in terms of I will solve the problem

Product: An outcome. The homework assignment you need to finish

The product can trigger your desire to procrasinate.

Focusing on process rather than product helps you not judge yourself (am I getting closer to finishing?)

Limiting distractions also can help

**Harnessing Your Habits to Help You**

Instead of using your willpower to change your entire habit, you can focus mainly on the Cue portion.

**Cues** as triggers can come from:

1. Location
2. Time
3. How you feel
4. Reactions

Having a plan for distractions during the **routine** helps remove distractions that lead to procrastination: Using the freedom app helps



Putting into Action:

Create weekly list of your key tasks in a planner journal

**Create a plan for the day before, NOT DAY OF**

Mix up your tasks

Planning your quitting time is as important as planning your working time

May want to look at something you are learning just before you go to sleep

Try to work on your most important and disliked tasks first (Eat your frogs in the morning)

The first few attempts may be challenging, but you should try not to get discouraged

## Week 4: How to Become a Better Learner

Sometimes in learning there are moments when you feel like you hit a wall. Language learners experience moments when what they learn seems incomprehensible. This can be explained by the brain restructuring and reorganizing what it has processed.



**Use Metaphor and Analogy:**

Metaphors/Analogies are very good to use for understanding concepts tying something unfamiliar with something familiar

The more visual you make metaphors/analogies, the better in terms of understanding

As you can better understanding where the metaphor becomes too simple or misleading, you can toss them away or create better more accurate ones

It can be helpful if you pretend that you are the concept you are trying to understand

Stories, even silly ones that are memory tricks without any context to the concept you are trying to learn, can help with your learning

Great scientific breakthroughs were made in chemistry when they created representations of molecules. Note, this is not what an actual molecule is or looks like, it is a abstract representation that allows understanding, as well as mutual communication about something if two people both have this grounding. **Testing**

You gain more from one hour taking a test than one hour studying the material

Checklist for Testing:

1. Did you make a serious effort to understand the text
2. Did you work with classmates on homework problems or at least check your solutions with others?
3. Did you attemp to outline every homework problem solution before working with classmates?
4. Did you actively in homework group discussions, contributing ideas and asking questions?
5. Did you consult with the instructor or teaching assistants when you were having trouble with something?
6. Did you understand all your homework problem solutions when they were handed in?
7. Did you ask in class for explanations of homework problem solutions that weren't clear to you?
8. If you had a study guide, did you carefully go through it before the test and convince yourself you could do everything on it?
9. Did you attempt to outline lots of problem solutions quickly without spending time on the algebra and calculations
10. Did you go for the study guide and problems with classmates and quiz one another?
11. If there was a review session before the test, did you attended and asked questions about anything you weren't sure about?
12. Did you get a reasonable night's sleep before the test?

**Hard to Easy Technique:**

(Note, this is not saying you should complete hard tests first)

When facing a test (or a problem that results in several tasks from easy to difficult):

1. Look at all the questions (try to understand the problem as a series that needs to be broken down)
2. Start first at what appears to be the hardest problem.
3. Pull away in within the first minute or so if you get stuck or don't feel like you are on the right track. This loads the difficult problem away from your mind and allows the diffuse mode to work on it.
4. Work on an easier problem, and continue making progress.

Analogy, working as a efficient chef, preparing multiple things at once to complete a meal.

Helpful Hints for Testing:

1. Try to change your mindset from anxiousness to excitedness at being challenged.

2. Turn your attention to your breathing. Relax your stomach, place your hand on it, slowly draw a deep breath. (diagraphmic breathing?)

3. Face your fears. Have a plan B. Once you have thought about a plan B, your fears can subside.

4. Disgintuish between good worry and bad worry: one helps focus, the other wastes energy.

5. Focusing on problems front to back, then checking them back to front sometimes allows fresh perspective.
